```python
import requests

from requests import get
from bs4 import BeautifulSoup
from warnings import warn
from time import sleep
from random import randint
import numpy as np, pandas as pd
import seaborn as sns
```

### Scrapping News Data

```python
kategori_count = {}
```

```python
berita_list = []
for page in range(600):
    url = "https://indeks.kompas.com/?page=" + str(page+1)
    # Sending a request to the speciifed URL
    response = requests.get(url)

    # Converting the responseonse to Beautiful Soup Object
    content = BeautifulSoup(response.content, 'lxml')

    if(page*15 % 150 == 0):
        print("Data ke- ", page*15)

    for berita in content.select('.article__list'):
           
        try:
            kategori_html = berita.select('.article__list__info')[0].select('.article__subtitle')[0].get_text()

            if(kategori_html not in kategori_count.keys()):
                kategori_count[kategori_html] = 0

            if(kategori_count[kategori_html] < 300):
                kategori_count[kategori_html] += 1

                response_isi_berita = requests.get(berita.select('.article__list__title')[0].select('a')[0].attrs.get('href'))
                isi_berita = BeautifulSoup(response_isi_berita.content, 'lxml')

                # Creating a python dictonary
                data = {
                    
                    "judul":berita.select('.article__list__title')[0].select('a')[0].get_text(),
                    "penulis":isi_berita.select('.read__credit__item')[0].select('a')[0].get_text(),
                    "tanggal_upload":berita.select('.article__list__info')[0].select('.article__date')[0].get_text().split(", ")[0],
                    "waktu_upload":berita.select('.article__list__info')[0].select('.article__date')[0].get_text().split(", ")[1],
                    "kategori":berita.select('.article__list__info')[0].select('.article__subtitle')[0].get_text(),
                    "body":isi_berita.select('meta')[41].attrs.get('content'),
                    "keywords":isi_berita.select('meta')[4].attrs.get('content')
                }
            else:
                continue
        except IndexError:
            print('Error')
            continue

        # print()
        # print('==============================================================================================================')
        # print(len(berita_list))
        berita_list.append(data)

df_berita = pd.DataFrame(berita_list)
print(df_berita.shape)
df_berita.head(5)
```
